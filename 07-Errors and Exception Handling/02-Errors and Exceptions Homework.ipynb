### Problem 1
Handle the exception thrown by the code below by using <code>try</code> and <code>except</code> blocks.

```python
for i in ['a','b','c']:
    print(i**2)
```

```python
for i in ['a','b','c']:
    try:
        print(i**2)
    except TypeError:
        print(f"{i} is an invalid number")
```
