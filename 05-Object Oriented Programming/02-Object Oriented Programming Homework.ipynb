#### Problem 1
Fill in the Line class methods to accept coordinates as a pair of tuples and return the slope and distance of the line.

```python
import cmath

class Line:
    
    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2
        
    
    def distance(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        
        dist = pow(x2-x1,2)+pow(y2-y1,2)
        return (dist**0.5)
        
    
    def slope(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        
        return ((y2-y1)/(x2-x1))
        
```

```python
# EXAMPLE OUTPUT

coordinate1 = (3,2)
coordinate2 = (8,10)

li = Line(coordinate1,coordinate2)
```

```python
li.distance()
```

```python
li.slope()
```

Fill in the class 

```python
class Cylinder:
    
    pi = 3.14
    
    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius = radius
        
    def volume(self):
        return Cylinder.pi*pow(self.radius,2)*self.height
    
    def surface_area(self):
        return (2*Cylinder.pi*self.radius*self.height)+(2*Cylinder.pi*pow(self.radius,2))
```

```python
# EXAMPLE OUTPUT
c = Cylinder(2,3)
```

```python
c.volume()
```

```python
c.surface_area()
```
