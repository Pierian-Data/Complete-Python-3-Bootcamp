# Object Oriented Programming Challenge

For this challenge, create a bank account class that has two attributes:

* owner
* balance

and two methods:

* deposit
* withdraw

As an added requirement, withdrawals may not exceed the available balance.

Instantiate your class, make several deposits and withdrawals, and test to make sure the account can't be overdrawn.

```python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def __str__(self):
        return f'Account owner: {self.owner}\nAccount balance: {self.balance}'        

    def deposit(self, dep_amt):
        self.balance += dep_amt
        return f'Funds Accepted! You have ${self.balance} in your account.'
    def withdraw(self, wthdrw_amt):
        if self.balance >= wthdrw_amt:
            self.balance -= wthdrw_amt
            return f'Withdrawal Accepted! You have ${self.balance} remaining in your account.'
        else:
            return 'Insufficient Funds!'
        
```

```python
# 1. Instantiate the class
acct1 = Account('Jose',100)
```

```python
# 2. Print the object
print(acct1)
```

```python
# 3. Show the account owner attribute
acct1.owner
```

```python
# 4. Show the account balance attribute
acct1.balance
```

```python
# 5. Make a series of deposits and withdrawals
acct1.deposit(50)
```

```python
acct1.withdraw(75)
```

```python
# 6. Make a withdrawal that exceeds the available balance
acct1.withdraw(500)
```
