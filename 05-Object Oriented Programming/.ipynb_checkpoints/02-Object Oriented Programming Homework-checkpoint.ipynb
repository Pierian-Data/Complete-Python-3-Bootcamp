#### Problem 1
Fill in the Line class methods to accept coordinates as a pair of tuples and return the slope and distance of the line.

```python
class Line:
    
    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2
        
        #Tuple unpacking
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        
        #Tuple indexing
        self.x = (coor2[0]-coor1[0])
        self.y = (coor2[1]-coor1[1])
    
    def distance(self): #distancia d=sqr((x2-x1)^2+(y2-y1)^2)

        return ((self.x)**2 + (self.y)**2)**0.5
    
    def slope(self): #pendiente m=(y2-y1)/(x2-x1)
        
        return (self.y/self.x)
```

```python
# EXAMPLE OUTPUT

coordinate1 = (3,2)
coordinate2 = (8,10)

li = Line(coordinate1,coordinate2)
```

```python
li.distance()
```

```python
li.slope()
```

#### Problem 2

Fill in the class 

```python
import math

class Cylinder:
    pi_math = math.pi
    pi = 3.14
    
    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius = radius
        
    def volume(self): #pi*r^2*h
        print ("Con la librería math el volumen es", self.pi_math*self.height*(self.radius)**2, " y con la asignacion de 3.14 a pi el volumen es", self.pi*self.height*(self.radius)**2)
    
    def surface_area(self): #2*pi*r*(h+r)
        return 2*self.pi*(self.radius)*(self.height+self.radius)
```

```python
# EXAMPLE OUTPUT
c = Cylinder(2,3)
```

```python
c.volume()
```

```python
c.surface_area()
```
